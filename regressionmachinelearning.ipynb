```python
#sales prediction based on advertising
#importing pandas libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading dataset
sales_data=pd.read_csv('https://msi.martial.co.ke/data/Advertising.csv')
sales_data.head()
```

```python
#data analysis
sales_data.info()
```

```python
sales_data.isnull().sum()
```

```python
sales_data.describe()
```

```python
#correlation analysis
sales_data.corr()
```

```python
#machine learning
#step 1: divide dataset into features(x) and label(y)
#convert sales dataset into numpy array to extract values
array=sales_data.values
array
#x=features(tv,radio,newspaper)
#200 records and 3 columns
x=array[:,0:3]
x
print(x.shape)

y=array[:,3]
print(y.shape)
```

```python
#step 2
#divide the x and y into training and testing
#x->x_train(70) and x_test(30)
#y->y-train(70) and y_test(30)

from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.30,random_state=42)
print (x_train.shape,y_train.shape)
print (x_test.shape,y_test.shape)
```

```python
#step 3
#import machine learning algorithm
#a. linear regression
from sklearn.linear_model import LinearRegression

#b. decision trees
from sklearn.tree import DecisionTreeRegressor

#c. k-nearest neighbor algorithm
from sklearn.neighbors import KNeighborsRegressor

#d. support vector machine
from sklearn.svm import SVR

#e. random forest
from sklearn.ensemble import RandomForestRegressor

#f. gradient boosting
from sklearn.ensemble import GradientBoostingRegressor

#g. adaboost
from sklearn.ensemble import AdaBoostRegressor

```

```python
#step 4: cross validating
#process that is used to determine the best learning algorithm for the particular problem
# STEP4: CROSS VALIDATION
# Hyperparameter Tuning.
# Process that is used to Determine the Best Learning Algorithm for a Particular Problem.
```

```python
#step 5: model training
#in this process, the models learnt from the training dataset using the best algorithm
model= DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
#step 6: model evaluation
# process used to determine whether model is accurate to predict the results
#in regression(numbers), we use the Mean Absolute Error(mae)

model_predictions=model.predict(x_test)
model_predictions
```

```python
y_test
```

```python
#scientific model evaluation
#regression we use r2_score
#ration between the model_predictions and y_test
from sklearn.metrics import r2_score
r2_score(y_test,model_predictions)*100
```

```python
#step 6: model prediction
#sales prediction for new data
#features=[300,50,120]
#outcome=model.predict(features)
#outcome
features = [[300, 50, 120]]
outcome = model.predict(features)
outcome
```
